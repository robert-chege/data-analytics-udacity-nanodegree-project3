```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import datetime as dt

%matplotlib inline
```

### Data Cleaning

#### Missing values in start_station_id , start_station_name, end_station_id, end_station_name, member_birth_year and member_gender

```python
# drope rows with missing values
bikedf = bikedf.dropna()
# tests all columns with missing values have been dropped
bikedf.info()
```

#### start_time and end_time are of the data type string instead of datetime

```python
# convert start_time and end_time data type to datetime
bikedf['start_time'] = pd.to_datetime(bikedf['start_time'])
bikedf['end_time'] = pd.to_datetime(bikedf['end_time'])
# test to confirm changes
bikedf[['start_time', 'end_time']].dtypes
```

#### bike_id datatype to string

```python
# convert bike_id data type to string
bikedf['bike_id'] = bikedf.bike_id.astype(str)
bikedf.dtypes
```

### Exploration and filtering the areas of interest

#### Extracting age of the bikers

```python
# extract year from start_time column
current_year = bikedf['start_time'].dt.year.unique()
print(current_year)

#  Calculate the bikers years and change the data type to int
bikedf['age'] = (current_year - bikedf['member_birth_year']).astype(int)

# test to confirm changes made
bikedf.age.info()
bikedf.head()
```

#### Dropping columns unnecessary for my analysis

```python
# columns to be dropped include: start_station_id,start_station_latitude, start_station_longitude, end_station_id,
#  end_station_latitude,end_station_longitude, bike_id , member_birth_year.
bikedf.drop(['start_station_id', 'start_station_latitude', 'start_station_longitude', 'end_station_id', 
        'end_station_latitude', 'end_station_longitude', 'bike_id' , 'member_birth_year'], axis = 1, inplace = True)
# test for all columns dropped
bikedf.info()
```

```python
bikedf.head()
```

```python
# To confirm different categories present in the dataset
print(bikedf['user_type'].unique())
print(bikedf['member_gender'].unique())
print(bikedf['bike_share_for_all_trip'].unique())
```

#### incorrect data_type: convert member_gender and user_type to category

```python
bikedf['member_gender'] = bikedf['member_gender'].astype('category')
bikedf['user_type'] = bikedf['user_type'].astype('category')
# Test
bikedf.info()
```

## Univariate Exploration


### What is the distribution of trip duration?

```python
bikedf.duration_sec.describe()
```

```python
# setting bins
binsize = 60
bins = np.arange(61, bikedf.duration_sec.max()+binsize, binsize)

# plotting
plt.figure(figsize=[8, 5])
plt.hist(data = bikedf, x = 'duration_sec', bins = bins);
plt.title('Distribution of Trip Durations', size = 15)
plt.xlabel('Duration (Sec)')
plt.ylabel('Number of Trips')
# set limits to get a better display of the distribution
plt.xlim([0, 3000]);
```

The plot is skewed to the right showing that trips that take longer are less frequent.

```python
# explore the distribution farther by changing xscale to log
np.log10(bikedf['duration_sec'].describe())
```

```python
# Set the scale to a log to get an even better outlook of the distribution
log_binsize = 0.05
bins = 10 ** np.arange(1.78, np.log10(bikedf['duration_sec'].max())+log_binsize, log_binsize)

plt.hist(data = bikedf, x = 'duration_sec', bins = bins)
plt.title('Distribution of Trip Durations', size = 15)
plt.xlabel('Duration (Sec)')
plt.ylabel('Number of Trips')
plt.xscale('log')
# set the appropriate ticks and labels
plt.xticks([50, 1e2, 5e2, 1e3, 5e3, 1e4, 2e4, 5e4, 1e5], [50, '100', '500', '1k', '5k', '10k', '20k', '50k', '100k']);
```

The most frequent trips take around 500 seconds.

### What is the distribution of member age?

```python
# gives the general overview of age in detail
bikedf.age.describe()
```

```python
# to get the total number of ages greater than 95 years
bikedf.age.loc[bikedf.age >= 95].sum()
```

It is highly unlikely to have ages above 95 years participating in bike rides. It is therefore logical to do away with this invalid data.

From the plot above, it is evident that the majority of users are aged 20-45 years with the peak being mid 30's. With increasing age the number of participants radically reduces, this is shown by right skew from the age of 50.

## Multivariate Exploration


```python
# reminder of columns to use in plotting
bikedf.columns
```

### What is the correlation of age with trip duration for each user type?

```python
g =sb.FacetGrid(data = bikedf, hue = 'user_type',
                hue_order = ['Subscriber', 'Customer'], aspect = 1.5)
g = g.map(sb.regplot, 'age', 'duration_min', x_jitter = 0.04, fit_reg = False)
g.add_legend();
plt.title('Correlation of Age with Trip Duration for Each User Type', size = 15)
plt.xlabel('Age (Years)')
plt.ylabel('Duration in Minutes (min)');
```

The correlation of trip duration for both subscribers appears to be similar.

### What is the correlation of age with trip duration for the bike shares?

```python
g =sb.FacetGrid(data = bikedf, hue = 'bike_share_for_all_trip',
                hue_order = ['No', 'Yes'], aspect = 1.5)
g = g.map(sb.regplot, 'age', 'duration_min', x_jitter = 0.04, fit_reg = False)
g.add_legend();
plt.title('Correlation of Age with Trip Duration for the Bike Shares', size = 15)
plt.xlabel('Age (Years)')
plt.ylabel('Duration in Minutes (min)');
```

The correlation of trip duration with age appears to be similar for both both bike shares. However, those who do not share their bikes appear to be taking the longer duration trips.

### What is the correlation of member age with trip duration for each gender?

```python
plt.figure(figsize=[10,15])
g = sb.FacetGrid(data = bikedf, col = 'member_gender', col_wrap = 3, height = 5,
                  aspect =0.65, xlim = [10, 95], ylim = [-100, 1500])
g.map_dataframe(sb.scatterplot, 'age', 'duration_min', alpha= None)
g.set_axis_labels('Age (year)', 'Duration (min)')
g.add_legend();
```

It is evident that across all genders, members between age 20 and 50 take longer time in their trips. The male and female plots are similar in terms of the reduction in duration time with age while for other gender there is no definite reduction in trip duration with age, only a specific age around 30 years seems to be having longer durations than the rest.

### Trip duration across gender and bike shares 

```python
plt.figure(figsize = [8,6])
sb.pointplot(data = bikedf, x ='member_gender', y = 'duration_min', hue = 'bike_share_for_all_trip',
           palette = 'colorblind', linestyles = '', dodge = 0.4)
plt.title('Trip Duration across gender and bike_share_for_all_trip', size = 15)
plt.ylabel('Average Trip Duration (Min)');
```

- The other gender is seen to spend more time in the trips, followed by female then male genders.
- Among the male and female genders, there's an almost equal level in trip duration for both bike share levels while for the other gender those who do not share bikes take significantly more time in their trips.

### Trip duration across gender and user types

```python
plt.figure(figsize = [8,6])
sb.pointplot(data = bikedf, x ='member_gender', y = 'duration_min', hue = 'user_type',
           palette = 'colorblind', linestyles = '', dodge = 0.4)
plt.title('Trip Duration across gender and customer type', size = 15)
plt.ylabel('Average Trip Duration (Min)');
```

- Among the other gender both customers and subsribers took the longest trip durations, followed by female members then male members.
- Customers in each gender averagely took the longest durations.
