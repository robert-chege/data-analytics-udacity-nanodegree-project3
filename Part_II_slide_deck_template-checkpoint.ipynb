## Dataset Overview

This dataset includes information about 174,877 individual bike rides made in a bike-sharing system covering the greater San Francisco Bay area. The properties of the individual bike rides include: trip duration in seconds and minutes, trip start and end time, start and end station names, bike shares, user type, gender and age of members.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
bikedf = pd.read_csv('fordgobike_updated_tripdata.csv')
bikedf.head()
```

## Frequency Of Bike Sharing Among User Types

The visualization below represents the number of user types involved in bike sharing. It is evident none of the customers is involved in bike sharing througout all the trips.

```python
# bike share number of users in each user type category.
plt.figure(figsize=[14.70, 8.27])
sb.countplot(data = bikedf, x = 'user_type', hue = 'bike_share_for_all_trip', order=bikedf.user_type.value_counts().index,
             palette = 'Blues');
plt.title('Number of User Types Involved In Bike Sharing', size = 15)
plt.xlabel('User Type')
plt.ylabel('Number of User Types');
```

## Correlation of Age With Trip Duration For Each Bike Share Category

From the visualization below it is clear that for both bike share categories, there is a negative correlation between member age and trip duration, however, it is key to note that majority of the members who shared their bikes though a small proportion compared to the other avoid longer duration trips.

```python
# bike shares distribution in the relationship of trip duration to age
g =sb.FacetGrid(data = bikedf, hue = 'bike_share_for_all_trip',
                hue_order = ['No', 'Yes'], aspect = 1.5)
g = g.map(sb.regplot, 'age', 'duration_min', x_jitter = 0.04, fit_reg = False)
g.add_legend();
plt.title('Correlation of Age With Trip Duration For Each Bike Share Category', size = 15)
plt.xlabel('Age (Years)')
plt.ylabel('Duration in Minutes (min)')
g.figure.set_size_inches(14.70, 8.27);
```

## Trip Durations Across Bike Share Categories For Each Gender

The below visualization illustrates how different gender spend time in their trips with regard to whether they are involved in bike sharing or not. It is evident that across all genders those not involved in bike sharing have the longest durations. It is also key to note that the other gender averagely spends more time in their trips, followed by females then males.<br>
It is also important to note that among the male and female genders, there's an almost equal level in trip duration for both bike share levels while for the other gender those who do not share bikes take significantly more time in their trips.
 

```python
# Trip duration for both bike share categories in different genders.
plt.figure(figsize=[14.70, 8.27])
sb.pointplot(data = bikedf, x ='member_gender', y = 'duration_min', hue = 'bike_share_for_all_trip',
           palette = 'colorblind', linestyles = '', dodge = 0.4)
plt.title('Trip Duration Across gender And Bike Share Categories', size = 15)
plt.xlabel('Member Gender')
plt.ylabel('Average Trip Duration (Min)');
```

## Trip Durations Across Bike User Types In Each Gender

The below visualization illustrates how different gender spend time in their trips with regard to whether they are customers or subscribers. It is evident that across all genders customers spend quite a significant time in their trips. The trend of the other gender having the longest durations, followed by female then male is also evident.

```python
plt.figure(figsize=[14.70, 8.27])
sb.pointplot(data = bikedf, x ='member_gender', y = 'duration_min', hue = 'user_type',
           palette = 'colorblind', linestyles = '', dodge = 0.4)
plt.title('Trip Duration Across Gender and User Types', size = 15)
plt.xlabel('Member Gender')
plt.ylabel('Average Trip Duration (Min)');
```
